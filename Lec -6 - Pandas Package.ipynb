## Handling Missing values:

Handling missing values is very important before working on the dataset

The places where we have'NaN' are the missing values.

1) Finding the missing values

```python
# True wherever there is null value
df.isna()
```

```python
# Counting missing value in every column
df.isna().sum()
```

```python
# Counting total null values:
df.isna().sum().sum()
```

2) Handling :
    1. Drop the null values
    2. Fill the null values

### Drop: 

```python
# drops the entire row in which any null value is present
df2 = df.dropna()
df2
```

```python
# drops the entire column in which any null value is present
df2=df.dropna(axis='columns')
df2
```

```python
# deletes the row with atleast threshold non null values
df2=df.dropna(thresh=7)
df2
# the row with less than 7 non-null values will be deleted.
```

```python
# deletes entire row only if all the elements in the row are null
df2=df.dropna(how='all')
df2
```

### Filling:
 
    Filling the null values requires domain knowledge.

There are three types:

##### 1) Filling with mean, constant value or a string

```python
df2= df.fillna(df.mean())
df2   ## The null values are filled by the mean value of that column
```

```python
df2 =df.fillna('Missing value')
df2
```

##### 2) Padding:
 
     This has two types:
            
            bfill/Backward filling- Fills the missing values with the values just below it.
            ffill/Forward filling- Fills the missing values with the values above below it.(default)

```python
# bfill :
df2=df.fillna(method='bfill')
df2
```

```python
# ffill :
df2=df.fillna(method='ffill')
df2
```

This method may not work in some cases if the null value is in  first row in ffill or last row in bfill. 

#### 3) Interpolates

This is a special dynamic method by pandas.

We do not require domain knowledge in this case and it will choose dynamically the appropiate method and value.

```python
df2=df.interpolate()
df2
```
